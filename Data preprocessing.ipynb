## ‚úÖ <strong>¬øQu√© es la preparaci√≥n de datos?</strong>
Preparaci√≥n o preprocesamiento de datos es el proceso de convertir los datos desde un formato en bruto (formato con problemas de calidad, inconsistencias) a otro formato, donde los datos estan listos para el an√°lisis.

Es fundamental en el flujo de trabajo de cualquier proyecto de an√°lisis de datos o de construcci√≥n de modelos de machine learning. Involucra varias tareas, como:

1Ô∏è‚É£ <b>Limpieza de datos:</b> Eliminar datos incompletos, incorrectos o inconsistentes, as√≠ como manejar valores at√≠picos o nulos.

2Ô∏è‚É£ <b>Transformaci√≥n de datos:</b> Reestructurar los datos para que est√©n en el formato adecuado para el an√°lisis o modelado. Esto puede implicar la normalizaci√≥n, la codificaci√≥n de variables categ√≥ricas, la generaci√≥n de nuevas caracter√≠sticas (features), etc.

3Ô∏è‚É£ <b>Integraci√≥n de datos:</b> Combinar datos de m√∫ltiples fuentes o tablas para crear un conjunto de datos completo y coherente.

4Ô∏è‚É£ <b>Reducci√≥n de dimensiones:</b> Reducir la cantidad de caracter√≠sticas o dimensiones del conjunto de datos para eliminar la redundancia o el ruido y mejorar la eficiencia del modelo.

5Ô∏è‚É£ <b>Divisi√≥n de datos</b>: Separar el conjunto de datos en subconjuntos (entrenamiento, validaci√≥n, prueba) para su posterior an√°lisis o modelado.

La preparaci√≥n de datos es esencial para asegurar que los modelos de machine learning se construyan con datos de alta calidad, lo que a su vez mejora la precisi√≥n y confiabilidad de los resultados obtenidos.

<br>

## ‚úÖ <strong>¬øPor qu√© es necesario preparar los datos?</strong>
‚û°Ô∏è Los datos originales pueden tener problemas:
1. Incompletos
2. Con errores
3. Inconsistentes

‚û°Ô∏è Las t√©cnicas de anal√≠tica pueden funcionar mal:
1. Recibe n√∫meros o texto
2. Valores en misma escala
3. No soporta at√≠picos

<br>

## ‚úÖ <strong>Proceso de preparaci√≥n de datos:</strong>
1Ô∏è‚É£ <b>Recopilaci√≥n:</b> obtener la mayor cantidad de datos posibles para el an√°lisis.

2Ô∏è‚É£ <b>Integraci√≥n:</b> unir los datos de dos o m√°s fuentes de datos.

3Ô∏è‚É£ <b>Entendimiento y visualizaci√≥n:</b> comprender los datos, aplicando estad√≠stica o visualizaciones.

4Ô∏è‚É£ <b>An√°lisis de calidad:</b> se establece cual es el nivel de calidad que tienen los datos, si hay datos faltantes, at√≠picos o inconsistentes.

5Ô∏è‚É£ <b>Limpieza:</b> se corrige los problemas que se identific√≥ previamente aplicando distintas t√©cnicas.

6Ô∏è‚É£ <b>Transformaci√≥n:</b> aplicar diferentes transformaciones para poner los datos de manera adecuada para la t√©cnica de anal√≠tica que se aplicar√°, ya sea en formato de num√©rico o texto.

###<strong> Paso 21. T√©cnicas de muestreo: </strong>

Si el objetivo fuera predecir la variable is_canceled se deberia analizar el balance de cada una de las clases, a continuaci√≥n se muestran en un
gr√°ficos:


```python
#Variable s—ñ la reserva fue cancelada o no
sns.countplot(data=df, x = 'is_canceled')
plt.show()
```

‚û°Ô∏è Es evidente que hay m√°s datos de una que de la otra, pudiera aplicarse una t√©cnicas de submuestreo para balancear las clases:


```python
from sklearn.utils import resample

# Contar las clases:
count_class_No, count_class_Yes = df["is_canceled"].value_counts()

#Dividir los dataframes por las clases:
df_class_No = df[df["is_canceled"] == 0]
df_class_Yes = df[df["is_canceled"] == 1]

#submuestrear la clase mayoritaria No:
no_downsampled = resample(df_class_No,
                      replace=False, # sample without replacement
                      n_samples = count_class_Yes, #Number of samples to generate
                      random_state = 27) #reproducible results
                      #combinar dataframes
df_sample = pd.concat([df_class_Yes, no_downsampled])
```

```python
sns.countplot(data=df_sample, x = 'is_canceled')
plt.show()
```
